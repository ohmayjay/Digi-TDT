```python
def rens_ord(text_0): 
    text_1 = text_0.replace("\n"," ") #erstatter linjeskift /n med mellemrum 
    text_2 = text_1.replace("."," ")### her og nedenfor: erstatter specielle tegn med blanktegn 
    text_3 = text_2.replace(","," ")
    text_4 = text_3.replace(":"," ")
    text_5 = text_4.replace("*"," ")
    text_6 = text_5.replace("–"," ")
    text_7 = text_6.replace("'"," ")
    text_8 = text_7.replace("”"," ")
    text_ren = text_8.replace("-"," ")
    text_lav = text_ren.lower()## Erstatter kapitaler med små bogstaver
    text_token = text_lav.split()## Splitter teksten ind i en liste efter mellemrum 
    return text_token
```

```python
def rens_sæt(text_0):
    text_1 = text_0.replace("\n"," ")#erstatter linjeskift /n med mellemrum
    text_2 = text_1.replace(","," ") #erstatter specielle tegn med blanktegn
    text_3 = text_2.replace(":"," ")
    text_4 = text_3.replace("*"," ")
    text_5 = text_4.replace("–"," ")
    text_6 = text_5.replace("'"," ")
    text_7 = text_6.replace("”"," ")
    text_ren = text_7.replace("-"," ")
    text_lav = text_ren.lower()
    text_token = text_lav.split(".") ## splitter teksten ind i en liste efter sætninger - punktummer 
    return text_token
```

```python
import os                      # os tillader os bl.a. finder filplaceringer på computeren
import numpy as np             # Numpy leverer noget af matematikken, der ligger under Pandas 
import pandas as pd            # Pandas tillader os at importere, oprette og manipulere data frames
from pandas import DataFrame   # Nogle libraries har under-biblioteker. Underbiblioteker importeres med from-kommandoen
```

```python
txt_0 = open("Dronningenstaler.txt", encoding = "latin-1") ## Kunne ikke læse utf8, men virker med latin-1. 
DrM50 = txt_0.read()
```

```python
ÅrListe=DrM50.split("Dronningens nytårstale") # Laver en liste der splitter ved Dronningens nytårstale
```

```python
type(ÅrListe) #Nu har vi en liste 
```

```python
len(ÅrListe) #Længden af listen er 51 fordi Dronningen har holdt 51 taler(?) 
```

```python
rensÅrOrd = [] #opretter tom liste over for hvert års tale, 
for år in ÅrListe: #For hvert års tale...
    rensÅrOrd.append(rens_ord(år)) #køres rens_ord-funktionen og tilføjes til listen. 
```

```python
rensÅrOrd
```

```python
# Nu tæller vi antal ord i hvert års tale 
```

```python
txtLen = [] #definerer ny liste 
for år in rensÅrOrd: #Looper kommandoen for hvert år i listen rensÅrOrd...
    txtLen.append(len(år)) #tilføj længden af hver enkelt tales elementer (antal ord) til den nye liste 
```

```python
txtLen
```

```python
# Nu tæller vi antal sætninger 
rensÅrSæt = [] #Definerer ny liste rensÅrSæt
for år in ÅrListe: # For hver tale i ÅrListe loopes kommandoen...
    rensÅrSæt.append(rens_sæt(år))#tilføj strings skilt ved sætningsgrænsen punktum til listen rensKapSæt  
```

```python
rensÅrSæt
```

```python
# Nu vil vi finde antal sætninger for hvert års tale 
```

```python
ÅrSætAntal = []#Ny liste ÅrSætAntal - 
for år in rensÅrSæt: #for hvert år i rensÅrSæt fra før loopes kommandoen...
    ÅrSætAntal.append(len(år)) # tilføj hvor mange (længden) af sætninger der er pr. kapitel til den nye liste      
```

```python
ÅrSætAntal #Viser en liste over antal sætninger i hvert års tale
```
